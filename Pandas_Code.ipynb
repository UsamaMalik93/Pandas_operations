# Pandas Merging Function  ===>> merge()

```python
df1=pd.DataFrame({'ID':[1,2,3,4,5,6]
                 ,'class':[9,10,11,12,13,15]})
df1
```

```python
df2=pd.DataFrame ({'ID': [1,2,3,4,5,7],
    'Name':['A','B','C','D','E','F']})
df2
```

```python
pd.merge(left=df2, right=df1,on='ID')#MERGE on same column that both have is ID
```

```python
#shortcut
pd.merge(df2,df1,on='ID')
```

```python
pd.merge(df2,df1,on='ID' ,how='left') #now you can see ID=7 also included
#how parameter means that all values of left dataframe will be included 
#irerespecive of no of indexes in left equal to right df
```

```python
pd.merge(df2,df1,on='ID' ,how='inner') #nner includes common indexes of both dataframes
```

```python
pd.merge(df2,df1,on='ID' ,how='outer')  #outer included all values of both dataframes
```

```python
pd.merge(df2,df1,on='ID' ,how='right') #no of indexes based on right dataframe
```

```python
pd.merge(df2,df1,on='ID' ,how='outer',indicator='True')  #outer included all values of both dataframes
#indicator parameter indicates which values are from left or right dataframe
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4,5],
                 'Class':[6,7,8,9,10]})
df1
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                 "B":[10,20,30]})
df2=pd.DataFrame({'C':[4,5,6],
                 'D':[40,50,60]})
display(df1,df2)
```

```python
#for joining see the following syntax
df1.join(df2)
```

```python
#For joining indexes should be same (by default they are same )   ---- 
#But if we set our own we should give sames indexes to both df's
df1=pd.DataFrame({'A':[1,2,3],
                 "B":[10,20,30]},index=['a','b','c'])
df2=pd.DataFrame({'C':[4,5,6],
                 'D':[40,50,60]},)
display(df1,df2)
```

```python
#now we we join u will get following result
df1.join(df2)
```

```python
#indexes shuold be same
df1=pd.DataFrame({'A':[1,2,3],
                 "B":[10,20,30]},index=['a','b','c'])
df2=pd.DataFrame({'C':[4,5,6],
                 'D':[40,50,60]},index=['a','b','c'])
display(df1,df2)
```

```python
df1.join(df2)
```

```python
#what if we have different number of indexes in  df's
df1=pd.DataFrame({'A':[1,2,3],
                 "B":[10,20,30]},index=['a','b','c'])
df2=pd.DataFrame({'C':[4,5],
                 'D':[40,50]},index=['a','b'])
display(df1,df2)
```

```python
df1.join(df2)#u can see NaN values....by default how is set to left(1st df)
```

```python
df1.join(df2,how='right')#values are set to according to right df (2nd one we used in parenthesis)
```

```python
df1.join(df2,how='inner')#intersection (inner) which values are not NaN
```

```python
df1.join(df2,how='outer')#union(outer)
```

```python
#What if we have column name same in both of df's  ===>> see
df1=pd.DataFrame({'A':[1,2,3],
                 "B":[10,20,30]})
df2=pd.DataFrame({'A':[4,5],#U can see both df's have A as same column name (so column names are overlapping)
                 'D':[40,50]})
display(df1,df2)
```

```python
df1.join(df2)#column names are overlapping so it will give error (don't execute it) !!!!
```

```python
df1.join(df2,lsuffix='_1')# suffix will resolve this error... lsuffix means suffix added to left df and vise versa (rsuffix)
```

```python
df1.join(df2, rsuffix='_1')
```
